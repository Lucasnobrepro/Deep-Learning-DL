```python
# Udando tensorflow==2.0.0

# Bibliotecas
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix

from tensorflow.keras.layers import Dense, Dropout
from tensorflow import keras
from tensorflow.keras import backend as k
from tensorflow.keras.models import Sequential 
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
# Carregando dados;
data = pd.read_csv('../Databases/entradas_breast.csv') # Features;
labels = pd.read_csv('../Databases/saidas_breast.csv') # Classes;
```

```python
# Função para criar uma rede neural;
def createNetwork():

    # k.clear_session()
    # Criando camadas
    model = Sequential([
                        Dense(units=16, activation="relu", kernel_initializer="random_uniform", input_dim=30),
                        Dropout(0.2),
                        Dense(units=16, activation="relu", kernel_initializer="random_uniform"),
                        Dropout(0.2),
                        Dense(units=1, activation="sigmoid", kernel_initializer="random_uniform")])
    # units= Numero de neuronios na camada, activation=Função de ativação, kenel_initializer=Iniciazação dos pesos, input_dim= Numero de entradas da camada inicial;

    # Comfigurando optimazer
    optimizer = keras.optimizers.Adam(lr = 0.001,      # Taxa de aprendizagem;
                                      decay = 0.0001,  # Muda o lr ao longo do tempo;
                                      clipvalue = 0.5) # Limites do gradiente de aprendizado;
    model.compile(optimizer=optimizer,
                  loss="binary_crossentropy",  # Erro do modelo para classificação binaria;
                  metrics=["binary_accuracy"]) # Metrica de avaliação;
    return model
```

```python
model = KerasClassifier(build_fn=createNetwork,
                        epochs=100,
                        batch_size=10)

result = cross_val_score(estimator=model,
                        X=data, y=labels,
                        cv=10,
                        scoring = 'accuracy')
```

```python
result = result.mean()
desvio = result.std()
```
