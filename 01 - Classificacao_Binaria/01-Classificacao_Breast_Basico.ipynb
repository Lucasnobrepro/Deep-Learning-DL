```python
# Udando tensorflow==2.0.0

# Bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

from tensorflow.keras.layers import Dense
from tensorflow import keras
from tensorflow.keras.models import Sequential 
```

```python
# Carregando dados;
data = pd.read_csv('../Databases/entradas_breast.csv') # Features;
labels = pd.read_csv('../Databases/saidas_breast.csv') # Classes;

# Dividindo dados de treino e de teste;
X_train, X_test, y_train, y_test = train_test_split(data,       
                                                    labels,
                                                    test_size=0.25) # 25% dos dados serÃ£o para teste;
```

```python
# Construindo rede neural;
# calculo de quantidade de neurronios para camada oculta: (Entradas + Saidas)/2, Exemplo: (30 + 1)/2 = 15.5, arredondando => 16 Neuronio na camada oculta;
model = Sequential([
               Dense(units=16, activation = 'relu', kernel_initializer = 'random_uniform', input_dim=30),
               Dense(units=16, activation = 'relu', kernel_initializer = 'random_uniform'),
               Dense(units=1, activation = 'sigmoid')])
```

```python
# Configurando a rede neural;
optimizer = keras.optimizers.Adam(lr = 0.001,     # Learning Rate;
                                 decay = 0.0001,  # Muda o lr ao longo do tempo;
                                 clipvalue = 0.5) 
 
model.compile(optimizer=optimizer,
              loss="binary_crossentropy",  # Erro do modelo para classificaÃ§Ã£o binaria;
              metrics=["binary_accuracy"]) # Metrica de avaliaÃ§Ã£o;

model.fit(X_train, y_train,  batch_size = 10, epochs = 100)
```

```python
# Verificando pesos das camadas pÃ³s treinamento;
weight_0 = model.layers[0].get_weights();
weight_1 = model.layers[1].get_weights();
weight_2 = model.layers[2].get_weights();
```

```python
# Prevendo com o modelo
pred = model.predict(X_test)
pred = (pred > 0.5)

# AvaliaÃ§Ã£o
acc =  accuracy_score(y_test, pred)
matriz = confusion_matrix(y_test, pred)
```
